## Transfert learning

Apprendre un réseau de neurones profond prend énormément de temps. Il est raisonnable d'adopter les deux stratégies suivantes :

* Figer des parties du réseaux de neurones, ne réapprend que les dernières couches
* Utiliser un réseau de neurones tels quel et s'en servir comme entrée d'un modèle de machine learning classique
